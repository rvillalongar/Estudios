```python
import numpy as np
import pandas as pd 
from sklearn_pandas import DataFrameMapper
df =pd.read_csv('dataset/Carseats.csv', index_col=0)
df.head()
```

### Vamos a simular que tenemos un dataset de entrenamiento y un dataset de Test.

```python
train = df.iloc[:-50, :].copy()
test  =df.iloc[-50:,:].copy()
test.loc[:, 'Sales'] = 0

test.head()
```

Vamos a prepocesar el Dataset. <br>
primero vamos a standarizar las columnas continuas y etiquetar los encode (las que son del tipo categoria)
Vamos a ingresar el tipo de preprocesamiento que deseamos ingresar a la clase DataFrameMapper y la vamos a solicitiar como DataFrame en vez de un numpy array cuando se ejecute la transformación.


```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
column_tuples = [
    (['CompPrice'], [StandardScaler()]),
    (['Income'], [StandardScaler()]),
    (['Advertising'], [StandardScaler()]),
    (['Population'], [StandardScaler()]),
    (['Price'], [StandardScaler()]),
    (['ShelveLoc'], [LabelEncoder()]),
    (['Age'], [StandardScaler()]),
    (['Education'], [StandardScaler()]),
    (['Urban'], [LabelEncoder()]), 
    (['US'], [LabelEncoder()])
]

mapper = DataFrameMapper(column_tuples, df_out=True)
```

```python
mapper_fit = mapper.fit(train)
train_processed = mapper.transform(train)
train_processed.head()
```

### Como se puede observar en la salidas se encuentra el dataframe de Inicio (con Sales incluido) y todas las variables que se preprocesaron 

```python
mapper.transform(test).head()
```

## Guardar y Reutilizar

Si en algun momento se desea crear el *mapper* y reutilizarlo en otro proceso o codigo. Este se puede almacenar y posteriormente leer

```python
import pickle
column_tuples = [
  #  (['Sales'], None),
    (['ShelveLoc'], [LabelEncoder()]),
    (['Urban'], [LabelEncoder()]), 
    (['US'], [LabelEncoder()])]

mapper = DataFrameMapper(column_tuples, default=StandardScaler())

mapper_fit = mapper.fit(train)

# Almacenar para Despues
pickle.dump(mapper_fit, open('models/fitted_mapper.pkl', 'wb'))

#leer, en otro proceso
mapper_fit = pickle.load(open('models/fitted_mapper.pkl','rb'))

test_processed = mapper.transform(test)
```
